# USER - USER based collaborative filtering

```python
# To initiat the Collaborative Filtering we started with Picking 2 books and the reviewers for those books
book_1, book_2 = "Wild Animus", "The Da Vinci Code"

# Getting all the reviewers for these books
book_1_reviewers = final_BookDB[final_BookDB.BookTitle == book_1].UserID
book_2_reviewers = final_BookDB[final_BookDB.BookTitle == book_2].UserID
pd.value_counts(book_1_reviewers)
pd.value_counts(book_2_reviewers)
```

### Check if there are any common reviewers

```python

common_reviewers = set(book_1_reviewers).intersection(book_2_reviewers)

print "%d people have reviewed these 2 books" % len(common_reviewers)

# Checking the table with only the common reviewers
list_common_reviewers = []

for i in common_reviewers:
    list_common_reviewers.append(i)

common_reviewers_only = final_BookDB[final_BookDB.UserID.isin(list_common_reviewers)]

common_reviewers_only.head(10)
```

```python
# Let's create a function that collect the reviews of our common reviewers
def get_book_reviews(title, common_reviewers):
    mask = (final_BookDB.UserID.isin(common_reviewers)) & (final_BookDB.BookTitle==title)
    reviews = final_BookDB[mask].sort('UserID')
    reviews = reviews[reviews.UserID.duplicated()==False]
    return reviews
```

# Calculate nearest neighbour correlation using Pearson Coefficient

```python
# Importing the scipy library to measure the pearson correlation coefficient
from scipy.stats.stats import pearsonr

def calculate_correlation(book1, book2):
    # We start by finding the common reviewers
    book_1_reviewers = final_BookDB[final_BookDB.BookTitle == book1].UserID
    book_2_reviewers = final_BookDB[final_BookDB.BookTitle == book2].UserID
    common_reviewers = set(book_1_reviewers).intersection(book_2_reviewers)

    # Then we look for the reviews given by common reviewers
    book_1_reviews = get_book_reviews(book1, common_reviewers)
    book_2_reviews = get_book_reviews(book2, common_reviewers)
    print(book_1_reviews.head(10))
    print(book_2_reviews.head(10))
    # Calculate the Pearson Correlation Score
    correlation_coefficient=pearsonr(book_1_reviews.BookRating, book_2_reviews.BookRating)[0]
    return correlation_coefficient

# Print the correlation score
calculate_correlation(book_1,book_2)

```

# Prepare Dataset for the most reviewed books and rank them

```python
# As I want to avoid dealing with a huge sparse matrix, I will only select the top most reviewed books for our example 

most_reviewed_books = pd.DataFrame({'Count' : final_BookDB.groupby(["BookTitle"]).size()})\
                                    .reset_index().sort_values(['Count'],ascending = False)

most_reviewed_books['Rank']=most_reviewed_books['Count'].rank(ascending=0)
most_reviewed_books['ISBN']=final_BookDB['ISBN']
most_reviewed_books
```

# Top 13 books recommended

```python
# Getting the list of the most reviewed book

top_books = []

for i in most_reviewed_books.BookTitle[0:13]:
    top_books.append(i)
    
top_books
```

# Top 10 recommendation based on your purchased book

```python
# Function which can help in finding the top recommendation for your book
def get_recs(my_book,num):
    import numpy as np
    recommends = []
    for b in top_books:
        if my_book==b:
            continue
        cor=calculate_correlation(my_book, b)
        if np.isnan(cor):
            continue
        else:
            recommends.append((b, cor))
            
    recommends.sort(key=lambda tup: tup[1], reverse=True)
    return recommends[:10]

```
